# Basic Implementation

# Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

# Dataset

<center><img src="iris.png" width="600" height="600" /></center>

```python
 from sklearn. datasets import load_iris

# Load the Iris dataset
iris = load_iris()

# Extract the input features (X) and target variable (y)
X = iris.data
y = iris.target
```

```python
X.shape, y.shape
```

# Label preparation (To categorical)

```python
 y_categorical = np.zeros((150, 3))

# Convert the target variable (y) into a categorical format
# Each row of y_categorical represents a sample, and each column represents a class
# y[i] represents the class label of the i-th sample
# Setting y_categorical[i, y[i]] to 1 indicates that the i-th sample belongs to the corresponding class
for i in range(150):
    y_categorical[i, y[i]] = 1
    
y_categorical
```

# Feature normalization

```python
n_feature = X.shape[1]

# Create an array with the same shape as X to store the normalized values
X_normalized = np.zeros_like(X)

# Normalize each feature of the input data X
# The loop iterates over each feature (column) of X
for i in range(n_feature):
    # Calculate the minimum and maximum values of the current feature
    Min = np.min(X[:, i])
    Max = np.max(X[:, i])
    
    # Normalize the current feature using min-max scaling
    # The normalized values are stored in the corresponding column of X_normalized
    X_normalized[:, i] = (X[:, i] - Min) / (Max - Min)
```

# Data Shuffleing

```python
index_=np.random.permutation(150)
index_
```

```python
# Randomly shuffle the indices of the data set
index_ = np.random.permutation(150)


# Create a new array, x_new, by using the shuffled indices to rearrange the rows of X_normalized
x_new = X_normalized[index_]


# Create a new array, y_new, by using the shuffled indices to rearrange the rows of y_categorical
y_new = y_categorical[index_]
```

```python
y_new
```

```python
# Create an array of indices from 0 to the number of rows in x_new
indexes = np.arange(x_new.shape[0])


# Shuffle the array of indices randomly
np.random.shuffle(indexes)


# Determine the index range for the training data (90% of the total data)
train_index = indexes[0: int(0.9 * x_new.shape[0])]


# Determine the index range for the testing data (10% of the total data)
test_index = indexes[int(0.9 * x_new.shape[0]):]


# Create a new array, x_train, by using the training indices to select the corresponding rows from x_new
x_train = x_new[train_index]


# Create a new array, x_test, by using the testing indices to select the corresponding rows from x_new
x_test = x_new[test_index]


# Create a new array, y_train, by using the training indices to select the corresponding rows from y_new
y_train = y_new[train_index]


# Create a new array, y_test, by using the testing indices to select the corresponding rows from y_new
y_test = y_new[test_index]
```

```python
print('X Train shape: ',x_train.shape)
print('Y Train shape: ',y_train.shape)
print('X Test shape: ',x_test.shape)
print('Y Test shape: ',y_test.shape)
```

## Model Training

```python
# Import the necessary module from the scikit-learn library
from sklearn import tree


# Create a new instance of the DecisionTreeClassifier class
clf = tree.DecisionTreeClassifier()


# Fit the model to the training data using the fit() method
model = clf.fit(x_train, y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test)
```

# SHAP

```python
import shap

# compute SHAP values
explainer = shap.Explainer(model, x_train)
shap_values = explainer(x_train)
```

# Feature Importance

##### Decision Tree

```python
# Import the necessary module from the scikit-learn library
from sklearn import tree

# Create a new instance of the DecisionTreeClassifier class
clf = tree.DecisionTreeClassifier()

# Fit the model to the training data using the fit() method
model = clf.fit(x_train, y_train)

# Get the feature importances from the fitted model
importances = model.feature_importances_

# Get the shape (size) of the importances array
importances.shape
```

```python
importances
```

```python
# Create a new list, X, by iterating over the range of the shape (size) of the importances array and appending a string 'F' followed by the index value incremented by 1
X = ['F' + str(i+1) for i in range(importances.shape[0])]

# Set the font family for the plot to Times New Roman
plt.rcParams["font.family"] = "Times New Roman"

# Create a new figure with a size of 4 inches by 3 inches and a dpi of 200
plt.figure(figsize=(4,3), dpi=200)

# Create a bar chart using the X list as the x-axis values and the importances array as the y-axis values, with a width of 0.4 for each bar
plt.bar(X, importances, width=0.4)

# Set the x-axis label to 'Feature'
plt.xlabel('Feature')

# Set the y-axis label to 'Score'
plt.ylabel('Score')

plt.show()
```

##### Random Forest

```python
# Import the necessary module from the scikit-learn library
from sklearn.ensemble import RandomForestClassifier

# Create a new instance of the RandomForestClassifier class with 200 estimators (trees)
clf = RandomForestClassifier(n_estimators=200)

# Fit the model to the training data using the fit() method
model = clf.fit(x_train, y_train)

# Get the feature importances from the fitted model
importances = model.feature_importances_

# Get the shape (size) of the importances array
importances.shape
```
