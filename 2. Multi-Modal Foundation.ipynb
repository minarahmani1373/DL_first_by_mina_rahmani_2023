```python
from segment_anything import sam_model_registry, SamAutomaticMaskGenerator, SamPredictor
import matplotlib.pyplot as plt
import numpy as np
import os
import torch
```

```python
image_path = 'E:/Deep Course/Weeks/W12/Data/1.tif'

image = plt.imread(image_path)

sam_checkpoint = "sam_vit_h_4b8939.pth"
model_type = "vit_h"

sam = sam_model_registry[model_type](checkpoint=sam_checkpoint)
sam.to(device="cpu")

mask_generator = SamAutomaticMaskGenerator(sam)

masks = mask_generator.generate(image)

        
```

```python
def return_thematic_map(anns):
    output_mask=np.zeros((256,256))
    for i in range(len(anns)):
        x,y=np.where(masks[i]['segmentation'])
        output_mask[x,y]=i+1

    return output_mask
```

```python
thematic_map=return_thematic_map(masks)
thematic_map.shape
```

```python
plt.figure(figsize=(5,5),dpi=200)

plt.subplot(1,2,1)
plt.imshow(image)
plt.xticks([])
plt.yticks([])
plt.title('RGB Image')

plt.subplot(1,2,2)
plt.imshow(thematic_map)
plt.xticks([])
plt.yticks([])
plt.title('Thematic Map')

plt.show()
```
