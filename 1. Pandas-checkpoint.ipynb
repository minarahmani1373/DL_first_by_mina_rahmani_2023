# Pandas

The pandas package is the most important tool at the disposal of Data Scientists and Analysts. The powerful machine learning and glamorous visualization tools may get all the attention, but pandas is the backbone of most data projects.

Pandas is built on top of the NumPy package, meaning a lot of the structure of NumPy is used or replicated in Pandas. Data in pandas is often used to feed statistical analysis in SciPy, plotting functions from Matplotlib, and machine learning algorithms in Scikit-learn.

Jupyter Notebooks give us the ability to execute code in a particular cell as opposed to running the entire file. This saves a lot of time when working with large datasets and complex transformations
