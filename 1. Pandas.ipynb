# Pandas

Pandas is a popular open-source data manipulation and analysis library for the Python programming language. It provides highly efficient and flexible data structures, namely the DataFrame and Series, which are designed to handle structured data in a tabular form.

The core of pandas revolves around the DataFrame, which can be thought of as a two-dimensional table or spreadsheet-like structure with labeled axes (rows and columns). It allows users to store, process, and analyze data in various formats such as CSV, Excel, SQL databases, and more. The DataFrame is capable of handling heterogeneous data types and missing values, making it suitable for real-world datasets that often contain inconsistencies.

One of the key features of pandas is its powerful data manipulation capabilities. It offers a wide range of functions and methods to filter, sort, transform, aggregate, and reshape data. This includes operations like merging, joining, pivoting, grouping, and more. These operations allow users to perform complex data manipulations with ease, enabling them to clean, preprocess, and transform data before analysis.

Pandas also provides intuitive indexing and slicing mechanisms, allowing users to access specific subsets of data efficiently. It supports both label-based and position-based indexing, making it easy to retrieve rows and columns based on their names or numerical positions. This flexibility facilitates data exploration and extraction, enabling users to extract insights from the data quickly.

Additionally, pandas integrates seamlessly with other Python libraries such as NumPy, Matplotlib, and scikit-learn, creating a powerful ecosystem for data analysis and machine learning. It enables users to leverage the functionalities of these libraries alongside pandas, further enhancing data analysis workflows.

Furthermore, pandas offers robust handling of time series data, including support for date/time indexing, resampling, frequency conversion, window operations, and more. This makes it a valuable tool for working with temporal datasets found in areas such as finance, economics, and scientific research.

In terms of performance, pandas is highly optimized and efficient. It is built on top of NumPy, which provides fast array manipulation capabilities. Pandas utilizes vectorized operations and avoids unnecessary loops to achieve high performance, especially when dealing with large datasets.

Overall, pandas has become the de facto standard for data analysis and manipulation in the Python ecosystem. Its rich set of features, ease of use, and extensive community support make it an indispensable tool for data scientists, analysts, and researchers working with structured data.

## Creating, Reading and Writing

### Creating data

There are two core objects in pandas: the DataFrame and the Series.

#### Dataframe

A DataFrame can be described as a tabular structure similar to a table. It consists of a collection of individual elements that hold specific values. Each element represents a combination of a row (or record) and a column in the DataFrame.

For example, consider the following simple DataFrame:

```python
#Create dataframe
dic={'name':['ali','Zahra','hossein'],
                  'sex':['male','female','male'],
                  'age':[23,25,19],
                  'address':[123,256,564]}

df = pd.DataFrame(dic,index=[12,13,15])
df
```

```python
df = pd.DataFrame()
df
```

```python
type(df)
```

```python
type(df[['name','age']])
```

```python
type(df['name'])
```

#### Series

A Series, on the other hand, refers to a collection of data values arranged in a specific order. While a DataFrame can be compared to a table, a Series can be likened to a list. In fact, you can easily create a Series by simply using a list.

```python
pd.Series([1, 2, 3, 4, 5])
```

```python
#series
df2=pd.Series([15,16,18,21,14],name='age2')
df2
```

Choosing specific values from a pandas DataFrame or Series is a crucial step in most data operations. Therefore, when working with data in Python, it is essential to learn how to select the relevant data points efficiently and promptly.

The second paradigm for attribute selection is the one followed by the loc operator: label-based selection. In this paradigm, it's the data index value, not its position, which matters

```python
data.loc[:, ['bedrooms', 'price', 'floors']]
```

## Conditional selection

So far we've been indexing various strides of data, using structural properties of the DataFrame itself. To do interesting things with the data, however, we often need to ask questions based on conditions.

For example, suppose that we're interested specifically in the houses having more than 3 floors

```python
data[data['floors']>3]
```

```python
data[data['floors']>2]
```

```python
data[data['floors']==2]
```

```python
data.loc[data.floors==2]
```

```python
data[(data['floors']==2) &(data['price']>data['price'].mean()) &(data['bathrooms']>2)]
```

```python
data['price'].mean()
```

```python
data[(data['floors']==2) &(data['price']>data['price'].mean()) | (data['bathrooms']>2)]
```

## NaN values

In pandas, NaN stands for "Not a Number" and it is used to represent missing or undefined values in a DataFrame or Series. NaN values can occur due to various reasons such as incomplete data, data corruption, or data transformation operations.

Here are some common operations related to handling NaN values in pandas:

```python
data.isnull()
```

```python
data.isnull().sum()
```

```python
data.fillna(0)
```

```python
data['bedrooms'].fillna(0)
```

To remove NaN values from a DataFrame in pandas, you can use the dropna() method. This method allows you to drop rows or columns containing NaN values based on your requirements

To remove rows with NaN values, you can call dropna() on your DataFrame as follows:

```python
data.dropna(axis=0, inplace=False)
```

In this example, the axis parameter is set to 0 to indicate that rows should be dropped. The inplace=True argument ensures that the changes are applied to the original DataFrame.

If you want to remove columns with NaN values instead, you can set axis to 1:

```python
data.dropna(axis=1, inplace=True)
data
```
