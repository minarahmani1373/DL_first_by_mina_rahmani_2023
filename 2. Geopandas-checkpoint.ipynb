# Geopandas

Geopandas is an open-source Python library that extends the functionalities of the popular pandas library to handle geospatial data. It provides a convenient and efficient way to work with geospatial datasets by combining the capabilities of pandas, shapely, and Fiona.

At its core, Geopandas allows users to read, write, manipulate, analyze, and visualize geospatial data in a tabular format. It introduces two new data structures: GeoSeries and GeoDataFrame. A GeoSeries is essentially a pandas Series object where each entry represents a geometric shape (a point, line, or polygon) with associated values. On the other hand, a GeoDataFrame is similar to a pandas DataFrame but includes a geometry column that stores the geospatial information.

One of the main advantages of Geopandas is its integration with the widely used libraries such as matplotlib for visualization and numpy for numerical computations. This makes it easy to perform spatial operations and analysis on geospatial data using familiar tools and techniques from the scientific Python ecosystem.

Geopandas leverages the underlying libraries to provide a rich set of spatial operations. These operations include geometric transformations (e.g., scaling, rotation), spatial joins (combining data based on spatial relationships), spatial queries (e.g., finding points within a certain distance of a polygon), and buffer operations (creating areas around geometries). These functionalities enable users to perform complex geospatial analysis without having to rely on external tools or specialized software.

Furthermore, Geopandas integrates well with other Python libraries dedicated to geospatial analysis, such as GeoPy, PySAL, and Rasterio. This allows users to leverage additional functionality for tasks like geocoding, spatial clustering, spatial autocorrelation analysis, and raster data processing seamlessly within their Geopandas workflows.

Another notable feature of Geopandas is its ability to read and write data in various common geospatial file formats, including shapefiles, GeoJSON, and GML. This makes it easy to import and export data to and from different sources, enabling interoperability with other GIS (Geographic Information System) software.

In summary, Geopandas is a powerful Python library that combines the tabular data handling capabilities of pandas with the spatial analysis and visualization functionalities of shapely and Fiona. It provides an accessible and efficient way to work with geospatial data, allowing users to perform various geospatial operations, analyze datasets, and create meaningful visualizations. Its integration with other popular Python libraries makes it a valuable tool for geospatial analysis within the Python ecosystem.

The main distinction between Pandas' data frames and Geopandas' GeoDataFrames lies in the presence of an extra column for geometries in GeoDataFrames. This column, called "geometry" by default, is a GeoSeries that stores the geometric objects such as points, lines, polygons, and more.

The fundamental operations like indexing, selecting, assigning, conditional selection, and others are very similar to those in the pandas package. However, Geopandas distinguishes itself by offering spatial operations and geometries specifically designed for working with spatial and vector data.

## Geometries in geopandas

Geopandas utilizes the geometric objects from the shapely library. These geometries are stored within a column named "geometry". To display the first 5 rows of this column, we can print them.

```python
data.geometry.head()
```

```python
print(data.geometry[0])
```

It contains geometrical data in the form of polygons and multipolygons with three-dimensional coordinates.

Each row represents a polygon or multipolygon object, and the "Z" suffix indicates that the coordinates include a third dimension (elevation or altitude). The numerical values inside the parentheses represent the X, Y, and Z coordinates of the vertices that define the shape.
